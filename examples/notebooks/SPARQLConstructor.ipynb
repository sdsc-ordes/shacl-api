```python
from SPARQLWrapper import SPARQLWrapper, BASIC, QueryResult
from rdflib import Graph, term, URIRef
from rdflib.namespace import Namespace, NamespaceManager

g = Graph()
db = SPARQLWrapper("http://imagingplazadev.epfl.ch:7300/repositories/imagingplaza")
query = '''
PREFIX ex: <http://example.com/ns#>
PREFIX schema: <http://schema.org/>
CONSTRUCT {?s ?p ?o} 
FROM NAMED <http://example.com/ns#id/> 
WHERE {graph <http://example.com/ns#id/> {?s ?p ?o .}}

'''
db.setHTTPAuth(BASIC)
db.setCredentials('adrien_coteries', 'shaclzodjsonsparql')
db.setQuery(query)
db.method = "GET"
db.setReturnFormat('json-ld')
db.queryType = "CONSTRUCT"
result = db.query()

jsonresult=(result._convertJSONLD())
# Print the converted result
# print(converted_result)
# print(result.response)
v = jsonresult.serialize(format='json-ld')
# v = jsonresult.serialize(format='turtle')
print(query)
print(v)

SCHEMA = Namespace("http://schema.org/")
SD = Namespace("https://w3id.org/okn/o/sd#")
BIO = Namespace("https://bioschemas.org/")
SPE = Namespace("https://openschemas.github.io/spec-container/specifications/")
RDF = Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
SH = Namespace("http://www.w3.org/ns/shacl#")
XSD = Namespace("http://www.w3.org/2001/XMLSchema#")
SHSH = Namespace("http://www.w3.org/ns/shacl-shacl#")
OWL = Namespace("http://www.w3.org/2002/07/owl#")
DCTERMS = Namespace("http://purl.org/dc/terms/")
EX = Namespace("http://example.com/ns#")
MD4I = Namespace("http://w3id.org/nfdi4ing/metadata4ing#")
RDFS = Namespace("http://www.w3.org/2000/01/rdf-schema#")

g.namespace_manager.bind('schema', SCHEMA, override=True, replace=True)
g.namespace_manager.bind('sd', SD, override=True, replace=True)
g.namespace_manager.bind('bio', BIO, override=True, replace=True)
g.namespace_manager.bind('spe', SPE, override=True, replace=True)
g.namespace_manager.bind('rdf', RDF, override=True, replace=True)
g.namespace_manager.bind('sh', SH, override=True, replace=True)
g.namespace_manager.bind('xsd', XSD, override=True, replace=True)
g.namespace_manager.bind('shsh', SHSH, override=True, replace=True)
g.namespace_manager.bind('owl', OWL, override=True, replace=True)
g.namespace_manager.bind('dcterms', DCTERMS, override=True, replace=True)
g.namespace_manager.bind('ex', EX, override=True, replace=True)
g.namespace_manager.bind('md4i', MD4I, override=True, replace=True)
g.namespace_manager.bind('rdfs', RDFS, override=True, replace=True)


g.parse(data=v, format='json-ld', Namespace=Namespace)

print(g.serialize(format='turtle'))
```
